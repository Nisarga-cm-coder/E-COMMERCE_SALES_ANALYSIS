```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('Netflix_Dataset_Updated (1).csv', encoding='unicode_escape', on_bad_lines='skip', engine='python')
```

```python
data['type'].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
data_sorted = data.sort_values(by='release_year')

# Create the gradient color palette
colors = sns.color_palette("coolwarm", len(data_sorted.head(20)))  # Adjust gradient range

# Create the bar plot
plt.figure(figsize=(12, 6))
bars = plt.bar(data_sorted['release_year'].head(20), data_sorted['total views'].head(20), color=colors)

# Labels and title
plt.xlabel("Release Year")
plt.ylabel("Total Views")
plt.title("Total Views per Release Year (Gradient)")

# Show the plot
plt.show()
```

```python
plt.subplot(2,1,1)
plt.bar(data['q1 views'],data['q2 views'])
plt.xlabel('Q1 views')
plt.ylabel('Q2 views')
plt.title('Q1 vs Q2 & Q3 vs Q4')

plt.subplot(2,1,2)
plt.bar(data['q3 views'],data['q4 views'])
plt.xlabel('Q3 views')
plt.ylabel('Q4 views')
```

```python
sns.set_style("whitegrid")

# Create subplots for Q1-Q4 views
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Define colors for each plot
colors = sns.color_palette("coolwarm", 4)

# Q1 Views
data.head(10).plot(kind='bar', x='q1 views', y='total views', ax=axes[0, 0], color=colors[0])
axes[0, 0].set_title("Q1 Views vs Total Views", fontsize=14)
axes[0, 0].set_xlabel("Q1 Views")
axes[0, 0].set_ylabel("Total Views")

# Q2 Views
data.head(20).plot(kind='bar', x='q2 views', y='total views', ax=axes[0, 1], color=colors[1])
axes[0, 1].set_title("Q2 Views vs Total Views", fontsize=14)
axes[0, 1].set_xlabel("Q2 Views")
axes[0, 1].set_ylabel("Total Views")

# Q3 Views
data.head(20).plot(kind='bar', x='q3 views', y='total views', ax=axes[1, 0], color=colors[2])
axes[1, 0].set_title("Q3 Views vs Total Views", fontsize=14)
axes[1, 0].set_xlabel("Q3 Views")
axes[1, 0].set_ylabel("Total Views")

# Q4 Views
data.head(20).plot(kind='bar', x='q4 views', y='total views', ax=axes[1, 1], color=colors[3])
axes[1, 1].set_title("Q4 Views vs Total Views", fontsize=14)
axes[1, 1].set_xlabel("Q4 Views")
axes[1, 1].set_ylabel("Total Views")

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
sns.set_style("whitegrid")

# Define colors
colors = sns.color_palette("viridis", 10)  # Using 'viridis' for better contrast

# Create figure
plt.figure(figsize=(12, 8))

# Plot with grouped barh
data.head(10).groupby(['release_year', 'type']).size().unstack().plot(
    kind='barh', stacked=True, color=colors, alpha=0.85, edgecolor="black"
)

# Improve visualization
plt.xlabel("Count", fontsize=12)
plt.ylabel("Release Year", fontsize=12)
plt.title("Content Type Distribution by Release Year", fontsize=14, fontweight='bold')
plt.legend(title="Type")
plt.grid(axis='x', linestyle="--", alpha=0.6)  # Light grid for better readability

# Show plot
plt.show()
```

```python
sns.set_style("whitegrid")

# Define colors
colors = sns.color_palette("magma", 4)  # Different colors for each quarter

# Group and sum quarterly views
quarterly_views = data.head(10)[['q1 views', 'q2 views', 'q3 views', 'q4 views']].sum()

# Create a horizontal bar plot
plt.figure(figsize=(12, 6))
quarterly_views.plot(kind='barh', color=colors, alpha=0.85, edgecolor="black")

# Improve visualization
plt.xlabel("Total Views", fontsize=12)
plt.ylabel("Quarters", fontsize=12)
plt.title("Total Views Per Quarter", fontsize=14, fontweight='bold')
plt.grid(axis='x', linestyle="--", alpha=0.6)

# Show plot
plt.show()
```

```python
sns.set_style("whitegrid")

# Define colors
colors = sns.color_palette("coolwarm", 4)  # Adjust number of colors dynamically

# Create figure
plt.figure(figsize=(12, 8))

# Group by 'star rating' and 'type', then plot
data.head(10).groupby(['star rating', 'type']).size().unstack().plot(
    kind='barh', stacked=True, color=colors, alpha=0.85, edgecolor="black"
)

# Improve visualization
plt.xlabel("Count", fontsize=12)
plt.ylabel("Star Rating", fontsize=12)
plt.title("Star Rating Distribution by Content Type", fontsize=14, fontweight='bold')
plt.legend(title="Type", fontsize=10)
plt.grid(axis='x', linestyle="--", alpha=0.6)

# Show plot
plt.show()
```

```python
import plotly.express as px
fig = px.scatter_3d(
    data,
    x='release_year',
    y='total views',
    z='total streaming duration',
    color='type',  # Different colors for content type
    size='total views',  # Size based on total views
    opacity=0.8,  # Slight transparency for better visualization
    title="3D Visualization of Streaming Data",
    labels={"release_year": "Release Year", "total views": "Total Views", "total streaming duration": "Streaming Duration"},
)

# Improve layout aesthetics
fig.update_layout(
    scene=dict(
        xaxis_title="Release Year",
        yaxis_title="Total Views",
        zaxis_title="Streaming Duration",
    ),
    template="plotly_dark",  # Dark theme for better contrast
    margin=dict(l=0, r=0, b=0, t=40)  # Reduce margins for a cleaner look
)

# Show plot
fig.show()
```
